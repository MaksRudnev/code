```python
browser = webdriver.Firefox()
browser.get('https://3ka.mipt.ru/admin/login/?next=/admin/')

#admin email
browser.find_element('xpath', '/html/body/div/div[2]/div/form/div[1]/input').send_keys('№№№№№№№№*****@phystech.edu')
#admin password
browser.find_element('xpath', '/html/body/div/div[2]/div/form/div[2]/input[1]').send_keys('Mangust№№№№№№*****')
#вход
browser.find_element('xpath', '/html/body/div/div[2]/div/form/div[3]/input').click()

#на сайте антидудос, чтобы не класть сайт
time.sleep(2.5)
#вкладка пользователи
browser.find_element('xpath', '/html/body/div/div[2]/div[1]/div[7]/table/tbody/tr/th/a').click()


#начало цикла
for i, row in users.iterrows():
    time.sleep(5)

    #на случай, если ошибка в заполнении, чтобы знать, на ком сломался цикл
    print(i)
    
    #добавить пользователя
    browser.find_element('xpath', '/html/body/div/div[3]/div/ul/li/a').click()

    time.sleep(2.5)
    #email
    browser.find_element('xpath', '/html/body/div/div[3]/div/form/div/fieldset/div[1]/div/input').send_keys(f"{row['Email']}")

    #password
    browser.find_element('xpath', '/html/body/div/div[3]/div/form/div/fieldset/div[2]/div/input').send_keys(f"{row['Password']}")

    #password2
    browser.find_element('xpath', '/html/body/div/div[3]/div/form/div/fieldset/div[3]/div/input').send_keys(f"{row['Password']}")

    time.sleep(2.5)
    #сохранить и продолжить
    browser.find_element('xpath', '/html/body/div/div[3]/div/form/div/div[3]/input[3]').click()

    #ФИО
    browser.find_element('xpath', '/html/body/div[1]/div[3]/div/form/div/fieldset[1]/div[1]/div/input').send_keys(f"{row['Фамилия']}")
    browser.find_element('xpath', '/html/body/div[1]/div[3]/div/form/div/fieldset[1]/div[2]/div/input').send_keys(f"{row['Имя']}")
    #browser.find_element('xpath', '/html/body/div[1]/div[3]/div/form/div/fieldset[1]/div[3]/div/input').send_keys('Тыгыдык')

    #ВК
    browser.find_element('xpath', '/html/body/div[1]/div[3]/div/form/div/fieldset[3]/div[1]/div/input').send_keys(f"{row['ВК']}")
    
    #комната
    browser.find_element('xpath', '/html/body/div[1]/div[3]/div/form/div/fieldset[4]/div[1]/div[2]/input').send_keys(f"{row['Комната']}")
    
    time.sleep(5)
    #сохранить
    browser.find_element('xpath', '/html/body/div[1]/div[3]/div/form/div/div[3]/input[1]').click()

#конец цикла
```
